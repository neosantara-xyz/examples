```python
from openai import OpenAI
import os
import time

client = OpenAI(
    api_key=os.getenv("NEOSANTARA_API_KEY"),
    base_url="https://api.neosantara.xyz/v1"
)
```

```python
print("Initiating video generation...")
response = client.post("/videos", json={
    "prompt": "A drone shot of a tropical island with crystal clear water",
    "model": "neosantara-video-v1"
})
video_id = response.json()["id"]

print(f"Video ID: {video_id}. Waiting for completion...")
while True:
    status_response = client.get(f"/videos/{video_id}")
    status = status_response.json()["status"]
    if status == "completed":
        print("Video generated successfully!")
        print(f"URL: {status_response.json()['url']}")
        break
    elif status == "failed":
        print("Video generation failed.")
        break
    time.sleep(10)
```
