```python
from openai import OpenAI
import os

client = OpenAI(
    api_key=os.getenv("NEOSANTARA_API_KEY"),
    base_url="https://api.neosantara.xyz/v1"
)
```

```python
response = client.chat.completions.create(
    model="glm-4.6",
    messages=[
        {
            "role": "user",
            "content": [
                {"type": "text", "text": "Extract merchant, date, and total from this receipt JSON."},
                {"type": "image_url", "image_url": {"url": "https://docs.neosantara.xyz/assets/receipt-example.jpg"}}
            ]
        }
    ],
    response_format={"type": "json_object"}
)
print(response.choices[0].message.content)
```
