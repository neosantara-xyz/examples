```python
from openai import OpenAI
import os

client = OpenAI(
    api_key=os.getenv("NEOSANTARA_API_KEY"),
    base_url="https://api.neosantara.xyz/v1"
)
```

```python
text_with_pii = "Nama saya Budi Santoso, nomor KTP saya 3171012345678901 dan email saya budi@example.com."

print("Original text:", text_with_pii)

# Using Nusantara-Base with guardrails enabled
response = client.chat.completions.create(
    model="nusantara-base",
    messages=[
        {"role": "system", "content": "Redact all PII (Name, KTP, Email) from the following text."},
        {"role": "user", "content": text_with_pii}
    ],
    extra_body={"guardrails": {"pii_redaction": True}}
)

print("Redacted text:", response.choices[0].message.content)
```
