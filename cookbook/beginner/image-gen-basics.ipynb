This notebook demonstrates how to generate images using the Neosantara AI API. You will learn how to use the `neosantara-gen-2045` model to create high-quality images from text prompts.

```python
# Install the OpenAI SDK
!pip install openai

# Setup API Key (for Google Colab)
try:
    from google.colab import userdata
    import os
    os.environ["NEOSANTARA_API_KEY"] = userdata.get('NEOSANTARA_API_KEY')
    print("API Key loaded from Google Colab userdata.")
except ImportError:
    import os
    if "NEOSANTARA_API_KEY" in os.environ:
        print("API Key loaded from environment variables.")
    else:
        print("Not running in Google Colab and NEOSANTARA_API_KEY not found in environment.")
```

```python
from openai import OpenAI
import os

client = OpenAI(
    api_key=os.getenv("NEOSANTARA_API_KEY"),
    base_url="https://api.neosantara.xyz/v1"
)
```

```python
from IPython.display import Image, display

response = client.images.generate(
    model="neosantara-gen-2045",
    prompt="A futuristic city in Nusantara, digital art style, vibrant colors",
    n=1,
    size="1024x1024"
)

image_url = response.data[0].url
print(f"Image URL: {image_url}")

# Display the image
display(Image(url=image_url))
```
