```python
from openai import OpenAI
import os

client = OpenAI(
    api_key=os.getenv("NEOSANTARA_API_KEY"),
    base_url="https://api.neosantara.xyz/v1"
)
```

```python
response = client.images.generate(
    model="neosantara-gen-2045",
    prompt="A futuristic view of Jakarta in 2045, hyper-realistic, digital art",
    n=1,
    size="1024x1024"
)

from IPython.display import Image, display
display(Image(url=response.data[0].url))
```
