```python
# Install the OpenAI SDK
!pip install openai

# Setup API Key (for Google Colab)
import os
try:
    from google.colab import userdata
    os.environ["NEOSANTARA_API_KEY"] = userdata.get('NEOSANTARA_API_KEY')
    print("✅ API Key loaded from Google Colab Secrets!")
except ImportError:
    if "NEOSANTARA_API_KEY" in os.environ:
        print("✅ API Key loaded from environment variables.")
    else:
        print("❌ NEOSANTARA_API_KEY not found!")
```

```python
from openai import OpenAI
import os

client = OpenAI(
    api_key=os.getenv("NEOSANTARA_API_KEY"),
    base_url="https://api.neosantara.xyz/v1"
)
```

```python
response = client.chat.completions.create(
    model="gemini-3-flash-preview",
    messages=[
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "user", "content": "Hello! What is Neosantara AI?"}
    ]
)

print(response.choices[0].message.content)
```
