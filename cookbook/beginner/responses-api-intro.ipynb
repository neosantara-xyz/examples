```python
from openai import OpenAI
import os

client = OpenAI(
    api_key=os.getenv("NEOSANTARA_API_KEY"),
    base_url="https://api.neosantara.xyz/v1"
)
```

### 🚀 Basic Usage: `client.responses.create()`

The `responses.create` method uses an `input` parameter instead of `messages`. It also provides a convenience property `output_text` to quickly access the model's final answer without iterating through chunks or tool outputs.

```python
response = client.responses.create(
    model="nusantara-base",
    input="Explain the concept of 'Gotong Royong' in Indonesian culture."
)

print(f"Response ID: {response.id}")
print("---")
print(response.output_text)
```
